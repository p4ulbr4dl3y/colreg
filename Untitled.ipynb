```python
from pipeline import VideoAnalyticsPipeline, draw_results
import cv2

pipeline = VideoAnalyticsPipeline()

# Загрузить изображения
ir_image = cv2.imread('ship.png')      # ИК-изображение (термальное)
visible_image = cv2.imread('ship.png')  # Видимое изображение (обычное)

# Обработать
result = pipeline.process_night(
    ir_image=ir_image,
    visible_image=visible_image,
    bbox_offset=(0, 0),      # Смещение между сенсорами (dx, dy), если нужно подогнать
    bbox_scale=(1.0, 1.5, 1.0, 1.0),  # Расширение: left, top, right, bottom
)

# Визуализация на ИК-изображении
vis = draw_results(ir_image, result)
cv2.imwrite('output_night.png', vis)

# Вывод результатов
for boat in result.boats:
    print(f"Boat {boat.boat_id}:")
    print(f"  Type: {boat.final_vessel_type}")
    print(f"  Binary classifier: {boat.vessel_type} ({boat.vessel_type_confidence:.1f}%)")
    if boat.lights_status:
        print(f"  Lights: {boat.lights_status.vessel_type}")
```
